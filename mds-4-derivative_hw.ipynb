Уровень 0

посчитайте значение 1-ой и 2-ой производной фунции
f(x) = x^5 + 4sin(2x) + cos(3x+3) в точке x = 1

```python
import numpy as np
from scipy.misc import derivative
import matplotlib.pyplot as plt
```

```python
def f(x):
    return x**5 + 4*np.sin(2*x) + np.cos(3*x+3)
```

```python
# точка
x=1
```

```python
# первая производная
derivative(f, x0=x, dx=1e-6, n=1)
```

```python
# вторая производная
derivative(f, x0=x, dx=1e-6, n=2)
```

```python
# график
x_axis = np.linspace(-2, 2)
#fx = f(x)
fx  = [f(xx) for xx in x_axis]
f1x = [derivative(f, x0=xx, dx=1e-6, n=1) for xx in x_axis]
f2x = [derivative(f, x0=xx, dx=1e-6, n=2) for xx in x_axis]


plt.plot(x_axis, fx, label='func')
plt.plot(x_axis, f1x, label='derivative 1')
plt.plot(x_axis, f2x, label='derivative 2')
plt.legend()
plt.show()
```

Постройте график первой и второй производной функции
(sin(2x+1))^5
на отрезке [-5; 5]

```python
def func(x):
    return (np.sin(2*x+1))**5
```

```python
# график
x_axis = np.linspace(-5, 5)
#fx = f(x)
fx  = [func(xx) for xx in x_axis]
f1x = [derivative(func, x0=xx, dx=1e-6, n=1) for xx in x_axis]
f2x = [derivative(func, x0=xx, dx=1e-6, n=2) for xx in x_axis]


plt.plot(x_axis, fx, label='func')
plt.plot(x_axis, f1x, label='derivative 1')
plt.plot(x_axis, f2x, label='derivative 2')
plt.legend()
plt.show()
```
